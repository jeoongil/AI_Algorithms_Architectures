```python
# 구글 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# 2. 라이브러리 import
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# 3. 경로 설정
train_dir = '/content/drive/MyDrive/my_dataset/train'
val_dir   = '/content/drive/MyDrive/my_dataset/validation'

# 4. 데이터 로드
train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(64, 64),
    batch_size=32,
    label_mode='categorical'
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    val_dir,
    image_size=(64, 64),
    batch_size=32,
    label_mode='categorical'
)

# 5. 클래스 이름 확인
class_names = train_dataset.class_names
print("클래스 목록:", class_names)

# 6. 정규화 및 prefetch
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.map(lambda x, y: (x / 255.0, y)).prefetch(buffer_size=AUTOTUNE)
val_dataset   = val_dataset.map(lambda x, y: (x / 255.0, y)).prefetch(buffer_size=AUTOTUNE)

# 7. 모델 정의
model = models.Sequential([
    layers.Input(shape=(64, 64, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')
])

# 8. 컴파일 및 학습
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_dataset, validation_data=val_dataset, epochs=10)

# 9. 시각화
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(6, 9))

plt.subplot(2, 1, 1)
plt.plot(epochs_range, acc, 'bo', label='Training acc')
plt.plot(epochs_range, val_acc, 'b', label='Validation acc')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(epochs_range, loss, 'bo', label='Training loss')
plt.plot(epochs_range, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 10. 결과 출력
print("Final Training Accuracy:", round(acc[-1], 4))
print("Final Validation Accuracy:", round(val_acc[-1], 4))
print("Final Training Loss:", round(loss[-1], 4))
print("Final Validation Loss:", round(val_loss[-1], 4))
```

```python
# 1. 구글 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# 2. 라이브러리 import (개선된 버전)
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# 3. GPU 메모리 최적화 (선택사항)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        tf.config.experimental.set_memory_growth(gpus[0], True)
    except RuntimeError as e:
        print(e)

# 4. 하이퍼파라미터 설정
IMG_HEIGHT, IMG_WIDTH = 224, 224  # 224x224로 변경 (전이학습 호환)
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 0.001

# 5. 경로 설정
train_dir = '/content/drive/MyDrive/my_dataset/train'
val_dir = '/content/drive/MyDrive/my_dataset/validation'

# 6. 개선된 데이터 증강
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,           # 회전 [4]
    width_shift_range=0.2,       # 좌우 이동 [4]
    height_shift_range=0.2,      # 상하 이동 [4]
    shear_range=0.2,            # 전단 변환 [4]
    zoom_range=0.2,             # 확대/축소 [4]
    horizontal_flip=True,        # 좌우 반전 [4]
    brightness_range=[0.8, 1.2], # 밝기 조절 [4]
    fill_mode='nearest',         # 빈 픽셀 채우기
    validation_split=0.2         # 검증 데이터 분할
)

val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# 7. 데이터 로드 (개선된 버전)
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = val_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# 8. 개선된 CNN 모델 (배치 정규화 + 드롭아웃)
def create_improved_cnn(num_classes):
    model = models.Sequential([
        # 입력층
        layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3)),

        # 첫 번째 Conv 블록
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),  # 배치 정규화로 학습 안정화 [9][12]
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),  # 과적합 방지 [9][12]

        # 두 번째 Conv 블록
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # 세 번째 Conv 블록
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # 네 번째 Conv 블록
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Global Average Pooling (매개변수 수 감소) [18][21]
        layers.GlobalAveragePooling2D(),

        # 완전연결층
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

# 9. 모델 생성
num_classes = len(train_generator.class_indices)
model = create_improved_cnn(num_classes)

# 10. 개선된 컴파일 설정
optimizer = tf.keras.optimizers.Adam(
    learning_rate=LEARNING_RATE,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07
)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 11. 콜백 함수 설정
callbacks_list = [
    # 조기 종료 [10]
    callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),

    # 학습률 감소 [13]
    callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),

    # 모델 체크포인트
    callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/best_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# 12. 모델 학습
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    callbacks=callbacks_list,
    verbose=1
)

# 13. 시각화 (요청하신 기존 코드 그대로 사용)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(6, 9))

plt.subplot(2, 1, 1)
plt.plot(epochs_range, acc, 'bo', label='Training acc')
plt.plot(epochs_range, val_acc, 'b', label='Validation acc')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(epochs_range, loss, 'bo', label='Training loss')
plt.plot(epochs_range, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 14. 결과 출력 (기존 코드 그대로 사용)
print("Final Training Accuracy:", round(acc[-1], 4))
print("Final Validation Accuracy:", round(val_acc[-1], 4))
print("Final Training Loss:", round(loss[-1], 4))
print("Final Validation Loss:", round(val_loss[-1], 4))

# 15. 추가 성능 평가
def evaluate_model_performance(model, validation_generator):
    # 예측 수행
    predictions = model.predict(validation_generator)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = validation_generator.classes

    # 정확도 계산
    accuracy = np.mean(predicted_classes == true_classes)

    # 클래스별 성능 계산
    class_names = list(validation_generator.class_indices.keys())

    print("\n=== 모델 성능 평가 ===")
    print(f"전체 정확도: {accuracy:.4f}")
    print("\n클래스별 상세 성능:")
    print(classification_report(true_classes, predicted_classes,
                              target_names=class_names, digits=4))

# 성능 평가 실행
evaluate_model_performance(model, validation_generator)
```

```python
# 1. 구글 드라이브 마운트 및 라이브러리
from google.colab import drive
drive.mount('/content/drive')

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils import class_weight

# 2. 하이퍼파라미터
IMG_HEIGHT, IMG_WIDTH = 32, 32
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.001

# 3. 데이터 경로
train_dir = '/content/drive/MyDrive/my_dataset/train'
val_dir   = '/content/drive/MyDrive/my_dataset/validation'

# 4. 데이터 증강 및 로딩
train_generator = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
).flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

validation_generator = ImageDataGenerator(rescale=1./255).flow_from_directory(
    val_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

class_names = list(train_generator.class_indices.keys())
print("클래스 목록:", class_names)

# 5. 클래스 가중치 계산
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weights = dict(enumerate(class_weights))
print("Class Weights:", class_weights)

# 6. 모델 정의
def create_mini_cnn(num_classes):
    model = models.Sequential([
        layers.Input(shape=(32, 32, 3)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),
        layers.Dropout(0.3),
        layers.GlobalAveragePooling2D(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

model = create_mini_cnn(len(class_names))
model.compile(optimizer=tf.keras.optimizers.Adam(LEARNING_RATE),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 7. 모델 학습 (콜백 없이, 끝까지 학습)
print("모델 학습 시작...")
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    class_weight=class_weights,
    verbose=1
)

# 8. 시각화 (빨간색: training, 파란색: validation)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(6, 9))

plt.subplot(2, 1, 1)
plt.plot(epochs_range, acc, 'r-', label='Training Accuracy')
plt.plot(epochs_range, val_acc, 'b-', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(epochs_range, loss, 'r-', label='Training Loss')
plt.plot(epochs_range, val_loss, 'b-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 9. 최종 성능 출력
print("Final Training Accuracy:", round(acc[-1], 4))
print("Final Validation Accuracy:", round(val_acc[-1], 4))
print("Final Training Loss:", round(loss[-1], 4))
print("Final Validation Loss:", round(val_loss[-1], 4))
```
